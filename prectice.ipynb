Distribution plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips=sns.load_dataset("tips")
sns.histplot(tips['total_bill'], bins=30)
plt.show()
```

```python
sns.kdeplot(tips['total_bill'], shade=True)
plt.show()
```

```python
sns.rugplot(tips['total_bill'])
plt.show()
```

```python
sns.displot(tips['total_bill'], kde=True)
plt.show()
```

```python
sns.displot(tips['total_bill'],bins=30, kde=True)
plt.show()
```

```python
sns.ecdfplot(tips['total_bill'])
plt.show()
```

```python
sns.violinplot(x=tips['day'], y=tips['total_bill'])
plt.show()

```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.boxplot(x=tips['day'], y=tips['total_bill'])
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips,kind='hex')
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
plt.show()
```

CATEGORICAL PLOT

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python

df = {
    'Subject': ['Math', 'Math', 'Math', 'Science', 'Science', 'Science',
                'English', 'English', 'English'] * 2,
    'Marks': [85, 90, 78, 88, 76, 92, 80, 75, 85,
              60, 65, 70, 55, 58, 68, 72, 74, 69],
    'Gender': ['Male']*9 + ['Female']*9
}

# Create DataFrame
data = pd.DataFrame(df)
print(data)
```

```python
#Shows average marks per subject.
sns.barplot(x="Subject", y="Marks", data=data)
plt.show()
```

```python
#Shows how many students took each subject.
sns.countplot(x="Subject", data=data)
plt.show()
```

```python
#Displays individual student marks with slight random spread
sns.stripplot(x="Subject", y="Marks",data=data,jitter=True)
plt.title("Marks by Subject with Jitter")
plt.show()
```

```python
#Similar to stripplot, but no overlap, clearer.
sns.swarmplot(x="Subject", y="Marks", data=data)
plt.show()
```

```python
# Shows median, quartiles, and outliers of marks per subject.
sns.boxplot(x="Subject", y="Marks", data=data)
plt.show()
```

```python
#Combines distribution shape + boxplot summary.
sns.violinplot(x="Subject", y="Marks", data=data)
plt.show()
```

```python
#Plots subject-wise mean marks with error bars.
sns.pointplot(x="Subject", y="Marks", data=data)
plt.show()
```

```python
#Creates a categorical plot (here same as stripplot, but flexible for grids).
sns.catplot(x="Subject", y="Marks", data=data, kind="strip")
plt.show()
```

```python
sns.catplot(x="Subject", y="Marks", data=data,kind="strip", jitter=True)
plt.show()
```
